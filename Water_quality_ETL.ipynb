## Extract

```python
import xml.etree.ElementTree as ET
import pandas as pd
```

```python
tree = ET.parse('C:\\Users\\Admin\\Downloads\\datafile.xml')
root = tree.getroot()
```

```python
print(root.tag)
print(len(root))
```

```python
data = []
for record in root:
    Year = record.find('YEAR').text
    State = record.find('STATE').text
    DO  = record.find('DO_MGL').text
    BOD = record.find('BOD_MGL').text
    FC  = record.find('FC_MPN100ML').text
    FS  = record.find('FS_MPN100ML').text
    pH = record.find('PH').text
    data.append({'Year':Year,'State':State,'DO (mg/L)':DO,'BOD (mg/L)':BOD,'FC (MPN/100ml)':FC,'FS (MPN/100ml)':FS,'pH':pH})
df1 = pd.DataFrame(data)
df1.head()
```

```python
main_df['DO (mg/L)'] = main_df['DO (mg/L)'].fillna(0)  # For numeric columns, use 0
main_df['BOD (mg/L)'] = main_df['BOD (mg/L)'].fillna(0)
main_df['FC (MPN/100ml)'] = main_df['FC (MPN/100ml)'].fillna(0)
main_df['FS (MPN/100ml)'] = main_df['FS (MPN/100ml)'].fillna(0)
main_df['pH'] = main_df['pH'].fillna(0)
```

```python
main_df = main_df.fillna(0)
```

```python
main_df.info()
```

```python
print(main_df["FC (MPN/100ml)"].str.contains('Na').sum())
```

```python
main_df["FC (MPN/100ml)"] = main_df["FC (MPN/100ml)"].replace(['-', 'NA'], np.nan)
main_df["FC (MPN/100ml)"] = pd.to_numeric(main_df["FC (MPN/100ml)"], errors='coerce')
main_df["FC (MPN/100ml)"] = main_df["FC (MPN/100ml)"].fillna(0)
```

```python
main_df["FS (MPN/100ml)"] = main_df["FS (MPN/100ml)"].replace(['-', 'NA'], np.nan)
main_df["FS (MPN/100ml)"] = pd.to_numeric(main_df["FS (MPN/100ml)"], errors='coerce')
main_df["FS (MPN/100ml)"] = main_df["FS (MPN/100ml)"].fillna(0)

main_df["pH"] = main_df["pH"].replace(['-', 'NA'], np.nan)
main_df["pH"] = pd.to_numeric(main_df["pH"], errors='coerce')
main_df["pH"] = main_df["pH"].fillna(0)
```

```python
main_df["DO (mg/L)"] = main_df["DO (mg/L)"].replace(['-', 'NA'], np.nan)
main_df["DO (mg/L)"] = pd.to_numeric(main_df["DO (mg/L)"], errors='coerce')
main_df["DO (mg/L)"] = main_df["DO (mg/L)"].fillna(0)

main_df["BOD (mg/L)"] = main_df["BOD (mg/L)"].replace(['-', 'NA'], np.nan)
main_df["BOD (mg/L)"] = pd.to_numeric(main_df["BOD (mg/L)"], errors='coerce')
main_df["BOD(mg/L)"] = main_df["BOD (mg/L)"].fillna(0)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

statewise_avg = main_df.groupby('State')['FC (MPN/100ml)', 'pH'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='State', y='FC (MPN/100ml)', data=statewise_avg)
plt.xticks(rotation=90)
plt.title('Average Turbidity by State')
plt.set_xlabel('State')
plt.set_ylabel('Fecal coliform')
plt.show()

# You can similarly plot for pH or Dissolved Oxygen:
plt.figure(figsize=(10, 6))
sns.barplot(x='State', y='pH', data=statewise_avg)
plt.xticks(rotation=90)
plt.title('Average pH by State')
plt.set_xlabel('State')
plt.set_ylabel('pH')
plt.show()
```

## Load

```python
import pyodbc
conn = pyodbc.connect('DRIVER={ODBC Driver 11 for SQL Server};'  
    'SERVER={server_name};'               
    'DATABASE=ETL_Project;'            
    'Trusted_Connection=yes;')
cursor = conn.cursor()
```

```python
create_table_query = '''
CREATE TABLE Water_Quality (
    Year INT,
    State VARCHAR(255),
    DO_mg_L FLOAT,
    BOD_mg_L FLOAT,
    FC_MPN_100ml FLOAT,
    FS_MPN_100ml FLOAT,
    pH FLOAT
);
'''
cursor.execute(create_table_query)
conn.commit() 
```

```python
import pandas as pd
for index, row in main_df.iterrows():
    insert_query = f'''
        INSERT INTO Water_Quality (Year, State, DO_mg_L, BOD_mg_L, FC_MPN_100ml, FS_MPN_100ml, pH)
        VALUES ({row['Year']}, '{row['State']}', {row['DO (mg/L)']}, {row['BOD (mg/L)']}, {row['FC (MPN/100ml)']}, {row['FS (MPN/100ml)']}, {row['pH']})
        '''
    cursor.execute(insert_query)
conn.commit()
```
